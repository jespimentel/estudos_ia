## PCA

- Variáveis categóricas: não são aceitas (podem ser transformadas em 0 e 1, por exemplo)
- Verificar se as unidades são as mesmas
- Se as unidades não forem as mesmas, não usar variância/covariância (isso só se usa se as variáveis forem padronizadas)
- Recomendação: correlação (usar Spearman em dados não normalizados)
- Sumária (summary)
- Eixos e variância explicadas por ele
- Scatter plot e biplot (com row labels)
- Pontos: são as observações
- Tamanho da seta: indica peso da variável
- Loadings
- Participação das variáveis nos eixos
- Use símbolos nos gráficos para as variáveis
- Duas variáveis que falam a mesma coisa: uma deve ser retirada (ex. variáveis que integram a fórmula das outras)
- 3 a 5 amostras para cada variável




## mtcars

```r
dados <- mtcars
```

```r
pca <- princomp(dados, cor = TRUE, scores = TRUE)
summary(pca)
```

```r
pca$loadings
```

```r
head(pca$scores)
```

```r
screeplot(pca) # mesmo resultado com plot(pca)
```

```r
x = pca$scores[,1]
y = pca$scores[,2]
plot(x, y, xlab = 'Componente principal 1', ylab = 'Componente principal 2')
```

```r
biplot(pca)
```

```r
library(devtools)
install_github("vqv/ggbiplot", force = TRUE)
```

```r
library(ggbiplot)
ggbiplot(pca)
```

```r
ggbiplot(pca, circle = TRUE)
```

# Análise de agrupamento

```r
mD <- dist(dados, method = 'euclidean')^2
```

```r
agr <- hclust(mD, method = 'complete')
```

```r
plot(agr, ylab="Dist. euclidiana quadrática", cex = 0.9)
```

```r
myGrupos <- cutree(agr, 3) 
print(myGrupos) 
```

```r
kM <- kmeans(dados, 3, nstart = 20)
print(kM$cluster)
```
